```python
# 建立新欄位資料
def Build_column(data,columnname):
    df = pd.DataFrame({ "Date":data["Date"],columnname:np.zeros(len(data["Date"]))})
    data= pd.merge(data,df)
    return data

# 新增均線欄位
# data請輸入股票欄位
def BuildWeightMeanline( data, daysmean ):
    long_num = len(data)
    nowColdata = [np.nan for x in range(daysmean-1)]

    for i in range(daysmean,long_num+1):
            # 計算每日的交易數量的權重平均
            tempsum = sum([data["Close"][j]*data["Volume"][j] for j in range(i-daysmean,i)])
            nowColdata.append( tempsum/sum(data["Volume"][i-daysmean:i]))

    print( len(nowColdata))
    data[ f"MA{daysmean}day_isweights"] = nowColdata
    return data
```

# 轉換成ticker

```python
def ticker_level(num):
  if num < 10:
    return 0.01
  elif num < 50:
    return 0.05
  elif num < 100:
    return 0.1
  elif num < 500:
    return 0.5
  elif num < 1000:
    return 1
  else:
    return 5 

def Compare_ticker(data ):
  data.columns = [str(x).lower() for x in data.columns ]
  ticker_pd = data[[x for x in data.columns if x not in ["date","volume", "index"] ]].copy()
  all_line = len(data)
  late_close = 0 
  for i in range(all_line):
    if late_close == 0 :
      late_close = data.iloc[ i,[x for x in data.columns].index("close")]
    else:
      ticker_num = ticker_level(late_close) 
      ticker_pd.iloc[i] = (ticker_pd.iloc[i] - late_close)/ticker_num
      # ticker_pd.iloc[i] = (ticker_pd.iloc[i] / late_close)-1
      late_close = data.iloc[ i,[x for x in data.columns].index("close")]
  ticker_pd["date"] = data["date"]
  ticker_pd["volume"] = data["volume"]  
  return ticker_pd

def Compare_percentage(data ):
  data.columns = [str(x).lower() for x in data.columns ]
  ticker_pd = data[[x for x in data.columns if x not in ["date","volume", "index"] ]].copy()
  all_line = len(data)
  late_close = 0 
  for i in range(all_line):
    if late_close == 0 :
      late_close = data.iloc[ i,[x for x in data.columns].index("close")]
    else:
      ticker_num = ticker_level(late_close) 
      # ticker_pd.iloc[i] = (ticker_pd.iloc[i] - late_close)/ticker_num
      ticker_pd.iloc[i] = (ticker_pd.iloc[i] / late_close)-1
      late_close = data.iloc[ i,[x for x in data.columns].index("close")]
  ticker_pd = ticker_pd.round(5)
  ticker_pd["date"] = data["date"]
  ticker_pd["volume"] = data["volume"]  
  return ticker_pd
```

```python

def ALL_Feature_processing(path):
  O_data = pd.read_csv(path)
  add_BBANDS = pd.merge(O_data,Build_BBANDS(O_data) ).copy()
  add_BBANDS = add_BBANDS.dropna()
  # to_ticker = Compare_ticker(add_BBANDS).copy().dropna()
  to_ticker = Compare_percentage(add_BBANDS).copy().dropna()
  O_data = O_data.iloc[to_ticker.index]
  process_end = pd.merge(to_ticker,Build_Kline(O_data) ).drop(0)
  return process_end

a = ALL_Feature_processing("/content/drive/My Drive/finance/TW0050.csv")
a
```

```python
def date_label(open,high,close ):
  label = list()
  npeye = np.eye(6)
  for i in range(len(open)):
    if open[i] > close[i]:
      if high[i] > open[i]:
        label.append(npeye[0])
      else:
        label.append(npeye[1])

    elif open[i] == close[i]:
      if high[i] > open[i]:
        label.append(npeye[2])
      else:
        label.append(npeye[3])
    else:
      if high[i] > open[i]:
        label.append(npeye[4])
      else:
        label.append(npeye[5])

  return label

# label = date_label(list(a.open),list(a.high),list(a.close))
```

```python
def Crate_file(path,daynum = 20):
  data_stock = list()
  label_out = list()
  data = ALL_Feature_processing(path)
  not_date = data[[x for x in data.columns if x not in ["date", "index"] ]].copy()
  not_date = not_date.dropna()
  label = date_label(list(not_date.open),list(not_date.high),list(not_date.close))
  for x in range(daynum,len(not_date)):
    data_stock.append(np.array(not_date[x-daynum:x]))
    label_out.append(label[x])

  data_stock = np.array(data_stock)
  label_out = np.array(label_out)
  return data_stock , label_out
```
